```python
import datetime as dt

from pytrends.request import TrendReq
```

```python
def create_timeframe_from_datetime(start_date: dt.date, end_date: dt.date):
    return str(start_date) + " " + str(end_date)
```

```python
def get_pytrends_data(kw_list, start, end, place=""):
    timeframe = create_timeframe_from_datetime(start, end)
    pytrends = TrendReq(hl='en-US', tz=360, retries=3, backoff_factor=0.1)
    pytrends.build_payload(kw_list, timeframe=timeframe, geo=place)
    data = pytrends.interest_over_time().drop(columns=["isPartial"])
    return data
```
