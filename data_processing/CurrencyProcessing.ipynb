```python
import pandas as pd
```

```python
def load_and_describe(path):
    currencies = pd.read_csv(path)
    
    print(f"Dataframe's shape: {currencies.shape}")
    print()
    
    print(f"Columns are: {[c for c in currencies.columns]}")
    print()
    
    print('Columns NAs:')
    nas = currencies.apply(lambda c : any(c.isna()), axis=0)
    print(f"NAs => {list(nas.loc[nas.to_numpy].index)}")
    print(f"No NAs => {list(nas.loc[(~nas).to_numpy].index)}")
    print()
    
    print(f'Types in df are \n{currencies.dtypes}')
    
    display(currencies.head())
    
    return currencies
```

# Loading currency.csv

```python
path = 'currency_usd.csv'
currencies = load_and_describe(path)
```

```python
# removing useless line
df_currencies = currencies.drop(index = 0)
# casting columns to right type
date_col = pd.to_datetime(df_currencies.Date, dayfirst=True)

currencies_col = [c for c in df_currencies.columns if 'Date' not in c]
df_currencies = pd.concat([df_currencies[c].apply(float) for c in currencies_col], axis=1)
df_currencies = pd.concat([date_col, df_currencies], axis=1)
display(df_currencies.dtypes[:5])
```

# Building time evolution for each currency

```python
df_history
```

```python
# Merging Ruble
_tmp = df_history[['Date', 'Dernier']].rename(columns={'Dernier':'RUB'})

RUB_x = df_currencies[['Date', 'RUB']]
RUB_y = _tmp[['Date', 'RUB']]
RUB = RUB_x.combine_first(RUB_y)
df_currencies = df_currencies.drop(columns='RUB').merge(RUB, on='Date', how='left')
df_currencies.head()
```

```python
df_currencies.sort_values('Date', inplace=True)
```

```python
tmp = df_currencies.Date.apply(lambda d : d.weekday()).values

tmp2 = []
previous_date = df_currencies.iloc[0].Date+pd.Timedelta(-1, 'D')
previous_day = tmp[0]-1

for day, date in zip(tmp, df_currencies.Date):
    diff = 0
    if (date - previous_date).days > 1 :
        diff = (date - previous_date).days - 1
        # print(f"DIFF {previous_day} {previous_date} -- {day} {date} -- {diff} ")
    
    if diff > 0:
        # print(list(range(1,diff+1)))
        for d in range(1,diff+1):  
            tmp2.append(((previous_day+d)%7, previous_date+pd.Timedelta(d, 'D'), date))
            # print(((day+d)%7, date+pd.Timedelta(d, 'D'), date))
            
    previous_day = day
    previous_date = date
    tmp2.append((day, date, date))

adjusted_dates = pd.DataFrame(tmp2).rename(columns={0:'dayofweek', 1:'actualDate', 2:'mergeDate'})
adjusted_dates.sort_values('actualDate').head()
```

```python
df_currencies = adjusted_dates.merge(df_currencies, left_on="mergeDate", right_on="Date").drop(columns=['mergeDate', 'dayofweek', 'Date']).rename(columns={'actualDate':'Date'})
df_currencies.sort_values("Date", ascending=False)
```

```python
def set_ref(df):
    _s = df.loc[df.Date==pd.to_datetime('2021-01-12')]
    ref = _s.iloc[:, 1].to_list()
    if len(ref) > 0:
        df['ref']=ref[0]
        return df
    else:
        raise Exception(f"{df.columns[1]} did not have value on the 2021-12-01")
```

```python
def set_min_max_avg_name(df):
    df['min'] = df.iloc[:, 1].min()
    df['max'] = df.iloc[:, 1].max()
    df['mean'] = df.iloc[:, 1].mean()
    results = []
    for c_id in currencies[df.columns[1]]:
        df['name'] = c_id[0]
        df['id'] = c_id[1]
        results.append(df.copy())
    return results
```

```python
def set_rel(df):
    ref = df.loc[:, 'ref'].values[0]
    df['rel(%)'] = df.iloc[:, 1]*100 / ref
    return df
```

```python
def set_curr(df):
    curr = df.columns[1]
    df['curr'] = curr
    df.rename(columns={curr:'value(vsUSD)'}, inplace=True)
    return df
```

```python
# Working per currency

data = list(df_currencies.columns[1:])
data = [df_currencies.loc[:, ['Date', c]] for c in data]
data = [set_ref(df) for df in data]
data = [df.loc[df.Date >= pd.to_datetime('2022-01-01')] for df in data]
data = [df.loc[df.Date <= pd.to_datetime('2022-05-26')] for df in data]
data = [df2 for df in data for df2 in set_min_max_avg_name(df)]
data = [set_rel(df) for df in data]
data = [set_curr(df) for df in data]

[display(df.head(3)) for df in data[27:30]]
```

```python
def set_decimal(df):
    for c in ['value(vsUSD)', 'ref', 'min', 'max', 'mean', 'rel(%)']:
        df[c] = df[c].round(3)
    return df

data = [set_decimal(df) for df in data]
[display(df.head(3)) for df in data[27:30]]
```
