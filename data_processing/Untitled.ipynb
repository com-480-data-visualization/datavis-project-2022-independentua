## Processing Sanctions data

```python
import pandas as pd
```

```python
data_loc = '../data/'
```

## Loading files 

```python
sanctions = pd.read_csv(data_loc + "NDA_Castellum_bycountry_bydate.csv")
sanctions.head()
```

```python
sanctions = sanctions.fillna(0.0).T
sanctions.head()
```

```python
columns = sanctions.iloc[0].values
print(columns)
rename_dict = dict(zip(range(len(columns)), columns))

sanctions = sanctions.iloc[1:].rename(columns=rename_dict)
sanctions.head()
```

```python
# pd.to_datetime(sanctions.reset_index()['index'])
tmp = sanctions.reset_index()['index'].str.split('.').apply(lambda x:x[0]+" 2022")
tmp = pd.to_datetime(tmp)
tmp[:10]
```

```python
sanctions = sanctions.reset_index(drop=True)
sanctions['date'] = tmp
sanctions.head()
```

```python
sanctions.to_csv(data_loc+"NDA_Castellum_bycountry_bydate_cleaned.csv", index=False)
```
